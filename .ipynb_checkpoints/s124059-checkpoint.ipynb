```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SequentialFeatureSelector
```


   #  Attribute                     Domain
   -- -----------------------------------------
       1. Sample code number            id number
       2. Clump Thickness               1 - 10

       4. Uniformity of Cell Shape      1 - 10
       5. Marginal Adhesion             1 - 10
       6. Single Epithelial Cell Size   1 - 10
       7. Bare Nuclei                   1 - 10
       8. Bland Chromatin               1 - 10

      10. Mitoses                       1 - 10
      11. Class:                        (2 for benign, 4 for malignant)

```python
columns = ['Sample code number','Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
```

```python
df = pd.read_csv('breast-cancer-wisconsin.csv',names=columns)
df
```

```python
df['Bare Nuclei'].str.isnumeric()
```

```python
df.head(30)
```

```python
df.describe()
```

```python
df.isnull().values.any()
```

```python
df = df.replace("?",np.NaN)
```

```python
df.head(30)
```

```python
df = df.dropna()
df.head(30)
```

```python
columns = df.columns
columns[4]
```

```python

colors = ListedColormap(['blue','red','black','magenta'])
plt.scatter(df[columns[7]],df[columns[9]],cmap=colors)
plt.xlabel(columns[7])
plt.ylabel(columns[9])
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y = Y.astype('category')
Y.value_counts()
```

```python
Y = Y.replace({2:0,4:1})
Y.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3 , random_state=42)
```

```python
x_test
```

```python
x_train
```

```python
KNN_model = KNeighborsClassifier(n_neighbors = 1)
```

```python
KNN_model.fit(x_train,y_train)
```

```python
y_predict = KNN_model.predict(x_test)
```

```python
accuracy_score(y_test,y_predict)
```

with feature Scalling (min_mix)

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
scaler.data_max_
```

```python
x_scaler_test = scaler.transform(x_test)
x_scaler_train = scaler.transform(x_train)
```

```python
KNN_model.fit(x_scaler_train , y_train)
```

```python
ysclaer_predict = KNN_model.predict(x_scaler_test)
```

```python
accuracy_score(y_test,ysclaer_predict)
```

feature selection

```python
KNN_model = KNeighborsClassifier(n_neighbors = 1)
sfs = SequentialFeatureSelector(KNN_model,n_features_to_select = 2)
sfs.fit(x_train, y_train)
X_train_reduced  = sfs.transform(x_train)
X_test_reduced = sfs.transform(x_test)
```

```python
sfs.support_
```

```python
KNN_model.fit(X_train_reduced,y_train)
y_pred_sfs = KNN_model.predict(X_test_reduced)
```

```python
accuracy_score(y_test,y_pred_sfs)
```
